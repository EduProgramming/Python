## 데이터 정보 확인

**1. info()**

info함수는 DataFrame 내의 컬럼명, 데이터 타입, Null건수, 데이터 건수 정보를 제공하여 줍니다.

```python
train.head()
```

```python
train.info()
```

**2. describe()**

describe는 데이터들의 평균, 표준편차, 4분위 분포도를 제공합니다.

```python
train.describe()
```

**3. value_counts()**

value_counts는 동일한 개별 데이터 값이 몇 건이 있는지 정보를 제공합니다.

⚠️ 단, value_counts는 Series객체에서 호출할 수 있으므로 DataFrame형태를 Series로 변환하여 호출하여야합니다.

```python
value_count = train['Pclass'].value_counts()
value_count
```
