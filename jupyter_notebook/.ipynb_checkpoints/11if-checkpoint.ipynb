## if ~ elif ~ elseë¬¸

ì¡°ê±´ì—ëŠ” 2ê°€ì§€ì˜ ê²½ìš°ê°€ ì•„ë‹Œ ë‹¤ì–‘í•œ ê²½ìš°ê°€ ì¡´ì¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.


````python
score = 78

if score >= 95:
    print("A+")
elif score >= 90:
    print("A")
elif score >= 85:
    print("B+")
elif score >= 80:
    print("B")
elif score >= 75:
    print("C+")
elif score >= 70:
    print("C")
elif score >= 65:
    print("D+")
elif score >= 60:
    print("D")
else:
    print("F")
````

ìœ„ì˜ ë‚´ìš©ì„ 2ì¤‘ ifë¬¸ì„ í†µí•´ì„œ ë‚˜íƒ€ë‚´ëŠ” ë™ì‹œì— ìš°ë¦¬ê°€ ë°°ì› ë˜ ë‚´ìš©ë“¤ì„ ì´ìš©í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
score = int(input("ì •ìˆ˜í˜• ìˆ«ìë§Œ ì…ë ¥ ê°€ëŠ¥í•©ë‹ˆë‹¤."))
scores = "ABCDF"
result = ""

if score >= 90:
    result = scores[0]
    if score >= 95:
        result += "+"
elif score >= 80:
    result = scores[1]
    if score >= 85:
        result += "+"
elif score >= 70:
    result = scores[2]
    if score >= 75:
        result += "+"
elif score >= 60:
    result = scores[3]
    if score >= 60:
        result += "+"
else:
    result = scores[-1]

print(result)
````

í˜ˆì•¡í˜•ì´ ë“¤ì–´ê°€ìˆëŠ” `blood`ë³€ìˆ˜ê°€ ìˆìŠµë‹ˆë‹¤. ì „ì‚°ì— ì˜¤ë¥˜ê°€ ìˆì—ˆëŠ”ì§€ í˜ˆì•¡í˜• í•˜ë‚˜ê°€ ë¹„ì–´ìˆì—ˆìŠµë‹ˆë‹¤. ì—†ëŠ” í˜ˆì•¡í˜•ì„ ì¶”ê°€í•˜ëŠ” ê°€ì •ë¬¸ì„ ì‘ì„±í•˜ì—¬ì£¼ì„¸ìš”. (ë¬¸ì œì˜ ì˜ˆì‹œë¡œëŠ” Oí˜•ì´ ì—†ëŠ” í˜•íƒœë¡œ ì‘ì„±í•˜ì˜€ìŠµë‹ˆë‹¤.)
